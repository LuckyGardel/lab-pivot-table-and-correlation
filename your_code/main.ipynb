#### What's the mean salary by department?

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# your answer here
def mean_no_max(x):
    no_max = []
    for i in x:
        if i == x.max():
            no_max 
        else:
            no_max.append(i)       
    mean = np.mean(no_max)
    return mean

df_mean_no_max = employee.pivot_table(index=['Department'],values= ['Salary'], aggfunc = ['mean',mean_no_max])
df_mean_no_max
```

After replacing the Nan Values:

```python
nan_rep2 = employee.pivot_table(index=['Department'],values= ['Salary'], aggfunc = ['mean',mean_no_max])
nan_rep2.fillna(55)
```
